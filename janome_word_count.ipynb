```python
for line in lines:
    # リスト内の1要素(文章)毎に形態素解析
    malist = t.tokenize(line)
    # 形態素解析した1つ1つをカウント
    for w in malist:
        # このとき、wには[単語、名詞、代名詞、、、]が格納されている
        # この格納されたものから「形態素された単語」、「品詞」を取得していく
        # surfaceは（分解された）形態素（意味がわかる最小限の言葉のかたまり）を指す
        # 参考:http://ailaby.com/janome/
        word = w.surface
        # part_of_speechに品詞情報が入っている
        part = w.part_of_speech
        # 名詞が含まれていない(0以下）なら処理しない
        if part.find('名詞') < 0:
            continue
        # 名詞だったら以下の処理に進む
        # if not:もし ○○ という条件を満たさない場合は、△△ という処理を行う
        # もし辞書になければカウンターを増やす
        if not word in worddic:
            # worddicのそのwordのカウントを0
            # 新しいキーを指定した場合は、そのキーと値を追加します。
            # ex:dict["white"] = 4  のように
            # dict = {'white': 4} となる
            ## 既に存在するキーを指定した場合は、そのキーの値を上書きする
            ## ex:dict["white"] = 5 のように
            ## dict = {'white': 5} となる
            # 参考:https://www.headboost.jp/python-adding-to-a-dict/
            worddic[word] = 0 # 初期化 今回wordは変数なのでクォーテーション必要なし
        # すでにカウンターがあった場合、wordのカウント(値?)を1つ増やす
        else:
            worddic[word] += 1
        #worddic[word] += 1　# なぜか数が1ずつ違う、、
```

```python
# 辞書型 worddic = {'銀河':27, '鉄道':6} といった{keys:values}という書き方をする
# worddic.keys()でkeysを、worddic.values()でvaluesを取得する
# 取得したものはリスト化され返される
# worddic.item()で(keys:values)の形を作成しリストを返す
# https://www.pythonweb.jp/tutorial/dictionary/index8.html
```

```python
# sorted(【リストor辞書】, key=lambda x: 【keyにしたい要素】)
# 基本sorted関数はこの形でしか使わないので覚えておく
# 参考:https://qiita.com/n10432/items/e0315979286ea9121d57

# sortedで辞書に入っているアイテムをxという名前で取り出して、その単語のカウント数を取り出す
#（x[1]とするとsortedはリストの2番目の要素、今回ならその単語のカウント数を返す）
# 因みにx[0]はリストの1番目の要素、今回は単語を返す
# そして、keyにしたものを基準にして降順にソートする、という処理をする
# sorted関数はリストを返す
# reverse=True:カウント数が大きい順
keys = sorted(worddic.items(), key = lambda x:x[1], reverse=True)
for word, cnt in keys[:50]:
    print('{0}({1})\n'.format(word, cnt), end='')
```
