```python
# mglearnのインストールができないのでirisに変更
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
print(iris.data.shape) # (150, 4)　150個のサンプルが４つの特徴量を保持
print(iris.feature_names) # 特徴量の名前['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
print(iris.target_names) # 品種名['setosa' 'versicolor' 'virginica']
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
df.corr()
```

```python
# データセットを並び替え、学習データとテストデータを分割
# テストデータを25%とし、乱数シードを固定
# コードが長いので2行にわたって記述
# よく使用する表現はtrain_test_split(train_X, train_y, test_size=0.25 ,random_state=0)という書き方
X_train, X_test, y_train, y_test = \
train_test_split(iris['data'], iris['target'], test_size=0.25 ,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier # k-最近傍法関数のimport
# モデル宣言
model = KNeighborsClassifier(n_neighbors=1)
```

```python
# モデル学習
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
print('Test set prediction:{}'.format(model.predict(X_test)))
```

```python
# モデルの汎化性能を評価
# scoreメソッドを使用
print('Test set accuracy:{}'.format(model.score(X_test, y_test)))
```
